```python
from datetime import datetime
# member_birth 컬럼을 datetime 객체로 변환
df_orders['member_birth_date'] = pd.to_datetime(df_orders['member_birth'], format='%Y.%m.%d')

# 현재 연도를 구하고, member_birth_date로부터 각 사용자의 나이를 계산
current_year = datetime.now().year
df_orders['member_age'] = current_year - df_orders['member_birth_date'].dt.year

# 연령대를 계산 (10년 단위로 구분)
df_orders['age_group'] = (df_orders['member_age'] // 10) * 10

# 결과 확인
df_orders[['member_birth', 'member_birth_date', 'member_age', 'age_group']]
```

```python
# 연령대와 성별에 따른 인기 제품 분석
popular_products_by_group = df_orders.groupby(['age_group', 'member_gender'])['product_id'].value_counts().groupby(level=[0, 1]).head(5)

def recommend_products(age_group, gender):
    recommendations = popular_products_by_group.loc[age_group, gender]
    return recommendations.index.tolist()

recommendations_20s_male = recommend_products(20, 1)
print(recommendations_20s_male)
```
